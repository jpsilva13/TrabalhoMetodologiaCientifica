```python
def quicksort(x):
    if len(x) == 1 or len(x) == 0:
        return x
    else:
        pivo = x[0]
        i = 0
        for j in range(len(x)-1):
            if x[j+1] < pivo:
                x[j+1],x[i+1] = x[i+1], x[j+1]
                i += 1
        x[0],x[i] = x[i],x[0]
        esq = quicksort(x[:i])
        dire = quicksort(x[i+1:])
        esq.append(x[i])
        return esq + dire
```

```python
def bubbleSort(x):
    for n in range(len(x)-1,0,-1):
        for i in range(n):
            if x[i]>x[i+1]:
                temp = x[i]
                x[i] = x[i+1]
                x[i+1] = temp
    return x
```

```python
def selection_sort(x):
    for i in range(len(x)-1):
        min_index = i
        for j in range(i+1, len(x)-1):
            if x[j] < x[min_index]:
                min_index = j
        x[i], x[min_index] = x[min_index], x[i]
```
